**If we have a close look at the 'size' columns unique values, the data is not in a number and we gonna take the number and store it in a 'bhk' as a new feature.**

```python
df2['bhk'] = df2['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3 = df2.copy()
df3['total_sqft'].unique()
```

```python
# Defining the function to check whether the element is float or not
def isfloat(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(isfloat)].head(15)
```

```python
# defining a function to convert the values of 'total_sqft' into float

def convert_to_float(x):
    tokens = x.split('-')
    if tokens == 2:
        return (token[0] + token[1]) / 2
# If it is number, turn it as a float
    try:
        return float(x)
    # If another format data comes, return false
    except:
        return None
    
```

```python
# Converting the values in feature 'total_sqft' to float 
df4 = df3.copy()
df4['total_sqft'] = df3['total_sqft'].apply(convert_to_float)
df4 = df4[df4.total_sqft.notnull()]
```

```python
# creating another feature called 'price_per_sqft' in the dataframe
df5 = df4.copy()
df5['price_per_sqft'] = (df4['price'] * 100000) / df4['total_sqft']
```

```python
df5.head()
```

```python
# length of unique values in feature 'location'
len(df5['location'].unique())
```

```python
# grouping the unique values in feature 'location' with their count (total number of datapoints for that location)
location_stats = df5.groupby('location')['location'].agg('count').sort_values(ascending = False)
location_stats
```

```python
# Displaying loactions which have less than 10 datapoints
location_stats[location_stats <= 10]
```

```python
# Locations with less than 10 datapoints
location_less_than_10_datapoints = location_stats[location_stats <= 10]
location_less_than_10_datapoints
```

```python
# Changing the feature 'location' values with 'others' which has less than 10 datapoints. 
df5['location'] = df5['location'].apply(lambda x: 'others' if x in location_less_than_10_datapoints else x)
df5.head(10)
```

**Here,if divide the tota_sqft with bhk, then it will give total sqft for 1 BHK.We are going to take only those whose value is greater than the threshold value 350. If the value is less than 350, it means it has 1BHK for that squarefeet which is normally an outlier. So, we are removing the rows whose total_sqft / bhk value is less than 350.**

```python
df5[df5.total_sqft / df5.bhk <350]
```

```python
# New dataframe without the outliers
df6 = df5[~(df5.total_sqft / df5.bhk <300)]
df6
```

```python
df6.price_per_sqft.describe()
```

```python
# Defining function to remove outliers for price_per_sqft
def remove_pps_outliers(df):
    df_new = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        mean = np.mean(subdf.price_per_sqft)
        sd = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft >(mean-sd)) & (subdf.price_per_sqft <= (mean+sd))]
        df_new = pd.concat([df_new, reduced_df], ignore_index = False)
    return df_new
```

```python
# New dataframe after removing outliers for price_per_sqft
df7 = remove_pps_outliers(df6)
df7
```
